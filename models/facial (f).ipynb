# 180 Seconds, Most Optimized. (250x250 Image Res)

```python
import os
import numpy as np
import face_recognition
import joblib
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from tqdm import tqdm
from PIL import Image

# Paths to dataset containing images
BRAD_PITT_PATH = "Brad"
OTHER_PEOPLE_PATH = "Other"

def process_image(img_path, label):
    try:
        image = face_recognition.load_image_file(img_path)
        
        # Resize image to 250x250 for faster processing
        pil_image = Image.fromarray(image)
        pil_image = pil_image.resize((250, 250))
        image = np.array(pil_image)
        
        face_encodings = face_recognition.face_encodings(image, model="small")
        if face_encodings:
            return face_encodings[0], label
    except Exception:
        pass
    return None

def load_images_and_labels():
    all_images = [(os.path.join(BRAD_PITT_PATH, img), "Brad Pitt") for img in os.listdir(BRAD_PITT_PATH)] + \
                 [(os.path.join(OTHER_PEOPLE_PATH, img), "Other") for img in os.listdir(OTHER_PEOPLE_PATH)]
    
    results = [process_image(img_path, label) for img_path, label in tqdm(all_images, desc="Processing Images")]
    valid_results = [res for res in results if res is not None]
    images, labels = zip(*valid_results) if valid_results else ([], [])
    
    return np.array(images), np.array(labels)

# Load dataset
encodings, labels = load_images_and_labels()

# Split dataset into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(encodings, labels, test_size=0.2, random_state=42)

# Train an SVM classifier with optimized hyperparameters
clf = SVC(kernel='linear', probability=True, max_iter=500)
clf.fit(X_train, y_train)

# Save the model
joblib.dump(clf, "face_recognition_model.pkl")

# Test the model
accuracy = clf.score(X_test, y_test)
print(f"Model Accuracy: {accuracy * 100:.2f}%")

#180 Seconds------------------
```

# Prediction Fn()

```python
def predict_image(image_path):
    image = face_recognition.load_image_file(image_path)
    face_encodings = face_recognition.face_encodings(image)
    
    if not face_encodings:
        print("No face found in the image!")
        return
    
    encoding = face_encodings[0]  # Take the first detected face
    prediction = clf.predict([encoding])
    probability = clf.predict_proba([encoding])

    print(f"Predicted Label: {prediction[0]}")
    print(f"Confidence: {max(probability[0]) * 100:.2f}%")
```

```python
import cv2
import face_recognition
import joblib
import numpy as np
import os
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Load or create model
MODEL_PATH = "face_recognition_model.pkl"
if os.path.exists(MODEL_PATH):
    clf = joblib.load(MODEL_PATH)
else:
    clf = SVC(kernel='linear', probability=True)

# Directory to store new face data
NEW_DATA_PATH = "NewFaces"
os.makedirs(NEW_DATA_PATH, exist_ok=True)

# Function to train model with new data
def train_model():
    images, labels = [], []
    for person in os.listdir(NEW_DATA_PATH):
        person_folder = os.path.join(NEW_DATA_PATH, person)
        for img_name in os.listdir(person_folder):
            img_path = os.path.join(person_folder, img_name)
            image = face_recognition.load_image_file(img_path)
            encodings = face_recognition.face_encodings(image)
            if encodings:
                images.append(encodings[0])
                labels.append(person)
    
    if images:
        X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
        clf.fit(X_train, y_train)
        joblib.dump(clf, MODEL_PATH)
        print("Model retrained with new data!")

# Initialize webcam
video_capture = cv2.VideoCapture(0)

while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    # Convert frame to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect faces
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations, model="small")

    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # Predict using trained model
        label = clf.predict([face_encoding])[0] if len(clf.classes_) > 0 else "Unknown"

        # Draw rectangle and label
        color = (0, 255, 0) if label != "Unknown" else (0, 0, 255)
        cv2.rectangle(frame, (left, top), (right, bottom), color, 2)
        cv2.putText(frame, label, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

        # Collect new face data
        print("Press 's' to save this face as a new entry.")
        if cv2.waitKey(1) & 0xFF == ord('s'):
            name = input("Enter person's name: ")
            person_folder = os.path.join(NEW_DATA_PATH, name)
            os.makedirs(person_folder, exist_ok=True)
            
            print(f"Collecting 50 images for {name}...")
            for i in range(50):
                ret, frame = video_capture.read()
                if not ret:
                    break
                img_path = os.path.join(person_folder, f"{name}_{i}.jpg")
                cv2.imwrite(img_path, frame[top:bottom, left:right])
                print(f"Saved: {img_path}")
            
            print(f"Collected 50 images for {name}. Training model...")
            train_model()  # Retrain model instantly

    # Show video feed
    cv2.imshow("Face Recognition", frame)

    # Quit on 'q' press
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Release webcam and close windows
video_capture.release()
cv2.destroyAllWindows()
```

# Model & Face Cam (INTEGRATED)

```python
import cv2
import face_recognition
import joblib
import numpy as np
import os
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

# Load or create model
MODEL_PATH = "ModelFaces.pkl"
if os.path.exists(MODEL_PATH):
    clf = joblib.load(MODEL_PATH)
else:
    clf = SVC(kernel='linear', probability=True)

# Directories to store face data
NEW_DATA_PATH = "NewFaces"
SECOND_DATA_PATH = "SecondDataset"
os.makedirs(NEW_DATA_PATH, exist_ok=True)
os.makedirs(SECOND_DATA_PATH, exist_ok=True)

# Function to train model with new data
def train_model():
    images, labels = [], []
    for dataset in [NEW_DATA_PATH, SECOND_DATA_PATH]:
        for person in os.listdir(dataset):
            person_folder = os.path.join(dataset, person)
            for img_name in os.listdir(person_folder):
                img_path = os.path.join(person_folder, img_name)
                image = face_recognition.load_image_file(img_path)
                encodings = face_recognition.face_encodings(image)
                if encodings:
                    images.append(encodings[0])
                    labels.append(person)
    
    if len(set(labels)) > 1:
        X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)
        clf.fit(X_train, y_train)
        joblib.dump(clf, MODEL_PATH)
        print("Model retrained with new data!")
    else:
        print("Error: At least two different classes are required to train the model.")

# Function to save images
def save_images(name, face_frames, dataset):
    person_folder = os.path.join(dataset, name)
    os.makedirs(person_folder, exist_ok=True)
    
    print(f"Saving {len(face_frames)} images for {name}...")
    for i, face_frame in enumerate(face_frames):
        img_path = os.path.join(person_folder, f"{name}_{i}.jpg")
        resized_face = cv2.resize(face_frame, (250, 250))  # Reduce size to save memory
        cv2.imwrite(img_path, resized_face)
    
    print(f"Saved {len(face_frames)} images for {name}. Training model...")
    train_model()

# Initialize webcam
video_capture = cv2.VideoCapture(0)
face_frames = []
face_name = ""
face_dataset = ""

while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    # Convert frame to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect faces
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations, model="small")

    for (top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        # Predict using trained model
        label = "Unknown"
        if hasattr(clf, "classes_") and len(clf.classes_) > 0:
            label = clf.predict([face_encoding])[0]

        # Draw rectangle and label
        color = (0, 255, 0) if label != "Unknown" else (0, 0, 255)
        cv2.rectangle(frame, (left, top), (right, bottom), color, 2)
        cv2.putText(frame, label, (left, top - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

        # Show options once per face detection
        if len(face_frames) == 0:
            print("\nPress 's' to start saving images in NewFaces.")
            print("Press 'd' to start saving images in SecondDataset.")
            print("Press 'q' to quit.")

        key = cv2.waitKey(1) & 0xFF
        
        if key == ord('s') or key == ord('d'):
            face_name = input("Enter person's name: ")
            face_dataset = NEW_DATA_PATH if key == ord('s') else SECOND_DATA_PATH
            face_frames = []

        if face_name and len(face_frames) < 50:
            face_frame = frame[top:bottom, left:right].copy()
            face_frames.append(face_frame)
            print(f"Collected {len(face_frames)} / 50 images...")

        if len(face_frames) == 50:
            save_images(face_name, face_frames, face_dataset)
            face_frames = []
            face_name = ""
            face_dataset = ""

    # Show video feed
    cv2.imshow("ModelFaces Recognition", frame)

    # Quit on 'q' press
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Release webcam and close windows
video_capture.release()
cv2.destroyAllWindows()
```
